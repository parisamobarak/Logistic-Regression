```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

df = pd.read_excel('dataset.xlsx', header=None)
df = df.values
print("dataset: \n", df)


x, y = df[:, :-1], df[:, -1]

# تبدیل Y به برچسب‌های دودویی (صفر و یک)
y_binary = np.where(y >= np.median(y), 1, 0)


x_train, x_test, y_train, y_test = train_test_split(x, y_binary, test_size=0.3, random_state=42)


mu = x_train.mean(axis=0)
sigma = x_train.std(axis=0)
xn_train = (x_train - mu) / sigma


xn_train = np.insert(xn_train, 0, 1, axis=1)


theta = np.random.randn(xn_train.shape[1])

# تابع لجستیک (فرضیه)
def h(x, theta):
    z = x.dot(theta)
    return 1 / (1 + np.exp(-np.clip(z, -500, 500)))

# تابع خطا لاگاریتمی
def log_loss(y_pred, y_true):
    epsilon = 1e-10  # برای جلوگیری از منفرج شدن لگاریتم صفر
    loss = -y_true * np.log(y_pred + epsilon) - (1 - y_true) * np.log(1 - y_pred + epsilon)
    return loss.mean()

# مرحله به روزرسانی نزول گرادیان
alpha = 0.001  # ضریب یادگیری
costs_train = []
costs_test = []

def update_step():
    global theta , costs_train
    y_pred= h(xn_train ,theta)
    costs_train.append(log_loss(y_pred , y_train)) 
    dtheta = np.zeros([xn_train.shape[0],xn_train.shape[1]])
    dtheta[:,0]=(y_pred - y_train)
    for i in range(1,xn_train.shape[1]):
        dtheta[:,i]= xn_train[: ,i]*dtheta[:,0]
        
    for i in range(0,xn_train.shape[1]):
        theta[i]-=alpha*dtheta[: ,i].mean()

costs_test=[]
for i in range(1000):
    update_step()
    x_test_n = (x_test - mu) / sigma  
    x_test_n = np.insert(x_test_n, 0, 1, axis=1)
    y_pred_test = h(x_test_n, theta)  
    costs_test.append((log_loss(y_pred_test, y_test)))
    #if i > 0 and (costs_train[-2] - costs_train[-1] < 0.1):
        #break  
#print(i) 

# رسم نمودار خطا در طول مراحل
plt.plot(costs_train, label='Training Set')
plt.plot(costs_test, label='Test Set ')
plt.xlabel('iteration')
plt.ylabel('costs')
plt.legend()
plt.show()

# تابع محاسبه دقت
def accuracy(theta, x, y):
    y_pred = h(x, theta)
    y_pred_class=np.where(y_pred >= 0.5, 1, 0)
    acc = np.mean(y_pred_class == y) * 100
    return acc

# محاسبه دقت برای مجموعه‌های آموزش و آزمون
train_accuracy = accuracy(theta, xn_train, y_train)
test_accuracy = accuracy(theta, x_test_n, y_test)

print(" train_accuracy: {:.2f}%".format(train_accuracy))
print(" test_accuracy: {:.2f}%".format(test_accuracy))
```
